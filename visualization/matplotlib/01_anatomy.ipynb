```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
title_lst = ['Training Image', '2 scales', '4 scales', '5 scales', '6 scales', '8 scales']
fig, axes = plt.subplots(2,3, figsize=(20,10), facecolor='linen')
for ax_index, ax in enumerate(axes.flat):
    ax.set_title(title_lst[ax_index], fontsize=20)
    ax.imshow(np.random.rand(100,100))
```

```python
fig, axes = plt.subplots(2,2, facecolor='linen', sharex=True, sharey=True)
```

```python
fig, axes = plt.subplots(3,1, figsize=(10,10), facecolor='linen', sharex=True)
fig.subplots_adjust(bottom=0.05, top=0.95, 
                    left=0.05, right=0.95, 
                    hspace=0.05)
axes[0].set_xlim([10,1000])
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,10), facecolor='linen', sharex=True, sharey=True)
fig.subplots_adjust(bottom=0.05, top=0.95, 
                    left=0.05, right=0.95, 
                    hspace=0.05)
axes[0,0].set_xlim([10,1000])
axes[0,0].set_ylim([10,1000])
```

```python
fig = plt.figure(figsize=(7,7))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212, sharex=ax1, sharey=ax1)
ax2.set_xlim([0,100])
```

```python
fig = plt.figure(figsize=(7,7))
ax1_1 = fig.add_subplot(221)
ax1_2 = fig.add_subplot(222, sharey=ax1_1)
ax2_1 = fig.add_subplot(223, sharex=ax1_1)
ax2_2 = fig.add_subplot(224, sharey=ax2_1, sharex=ax1_2)

ax1_1.set_ylim([-10,10])
ax2_1.set_ylim([0,100])
ax2_1.set_xlim([0,20])
ax2_2.set_xlim([20,40])

fig.tight_layout()
```
