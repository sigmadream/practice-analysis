```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(12,7), facecolor='linen')
ax = fig.add_subplot()
ax.plot([2,3,1])
ax.scatter([2,3,1],[2,3,4])
```

```python
fig = plt.figure(figsize=(12,7), facecolor='linen')
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)
ax1.plot([2,5,10])
ax2.plot([10,5,2])
ax3.scatter([10,5,2], [2,3,4])
```

```python
title_lst = ['Training Image', '2 scales', '4 scales', '5 scales', '6 scales', '8 scales']
fig, axes = plt.subplots(2,3, figsize=(20,10), facecolor='linen')
for ax_index, ax in enumerate(axes.flat):
    ax.set_title(title_lst[ax_index], fontsize=20)
    ax.imshow(np.random.rand(100,100))
```
